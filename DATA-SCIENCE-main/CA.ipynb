```python
import pandas as pd
import matplotlib.pyplot as plt

alpha = pd.read_csv("airlines_flights_data.csv") # This creates a dataframe from a csv
alpha.dropna(inplace=True)

alpha = alpha.rename(columns={
    "airline": "Airline",
    "flight": "Flight",
    "source_city": "Source_City",
    "departure_time": "Departure_time",
    "stops": "Stops",
    "arrival_time": "Arrival_time",
    "destination_city": "Destination_city",
    "class": "Class",
    "duration": "Duration",
    "days_left": "Days_left",
    "price": "Price",
})

numeric_cols = alpha.select_dtypes(include="number").columns
string_cols = alpha.select_dtypes(include="object").columns

all_cols = list(string_cols) + list(numeric_cols)

alpha = alpha.sort_values(by=all_cols, ascending=True)

import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
alpha = pd.read_csv("airlines_flights_data.csv")
alpha.dropna(inplace=True)
alpha = alpha.rename(columns={
    "airline": "Airline",
    "flight": "Flight",
    "source_city": "Source_City",
    "departure_time": "Departure_time",
    "stops": "Stops",
    "arrival_time": "Arrival_time",
    "destination_city": "Destination_city",
    "class": "Class",
    "duration": "Duration",
    "days_left": "Days_left",
    "price": "Price",
})

# 1. Bar Chart: Number of flights per Airline
flight_counts = alpha['Airline'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(flight_counts.index, flight_counts.values, color='purple')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.title('Number of Flights per Airline')
plt.xticks(rotation=45)
plt.show()

# 2. Scatter Plot: Days_left vs Price
plt.figure(figsize=(8,6))
plt.scatter(alpha['Days_left'], alpha['Price'], color='blue', alpha=0.6)
plt.xlabel('Days Left')
plt.ylabel('Price')
plt.title('Scatter Plot of Price vs Days Left')
plt.show()

# 3. Histogram: Price distribution
plt.figure(figsize=(8,6))
plt.hist(alpha['Price'], bins=20, color='green', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Number of Flights')
plt.title('Histogram of Flight Prices')
plt.show()

# 4. Line Plot: Average Price over Days_left
avg_price_per_day = alpha.groupby('Days_left')['Price'].mean()
plt.figure(figsize=(8,6))
plt.plot(avg_price_per_day.index, avg_price_per_day.values, marker='o', color='red')
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.title('Average Price vs Days Left')
plt.grid(True)
plt.show()

# 5. Pie Chart: Flight distribution by Airline
plt.figure(figsize=(8,8))
plt.pie(flight_counts.values, labels=flight_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Flight Distribution by Airline')
plt.show()
```
