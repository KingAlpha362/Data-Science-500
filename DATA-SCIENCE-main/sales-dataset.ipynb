```python
from matplotlib import pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data_sample.csv", encoding="latin-1")

# a. Display first 5 rows and shape
print("Shape of the data:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

# b. Rename columns
df = df.rename(columns={
    "QUANTITYORDERED": "Quantity Ordered",
    "PRICEEACH": "Price Each",
    "SALES": "Sales Amount"
})

# c. Add Performance column
def categorize_sales(amount):
    if amount < 1000:
        return 'Low'
    elif amount <= 4999:
        return 'Medium'
    else:
        return 'High'

df['Performance'] = df['Sales Amount'].apply(categorize_sales)

# d. Scatter plot: Quantity Ordered vs Sales Amount
plt.figure(figsize=(8,5))
plt.scatter(df['Quantity Ordered'], df['Sales Amount'], color='yellow', edgecolor='black')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales Amount')
plt.title('Quantity Ordered vs Sales Amount')
plt.show()

# e. Save the final dataset
df.to_csv("sales_data_with_performance.csv", index=False)
print("Final dataset saved as sales_data_with_performance.csv")
```
